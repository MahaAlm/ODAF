```python
def completeness():
    print("---------------------------------")
    print("_____مقياس اكتمال البيانات_____")
    completeness = 100.0
    filePath = input("🗎 الرجاء إدخال مسار الملف: ")
    if filePath.endswith('.xlsx'):
      # Load XLSX file
      data = pd.read_excel(filePath)
    elif filePath.endswith('.xls'):
      # Load XLSX file
      data = pd.read_excel(filePath)
    elif filePath.endswith('.csv'):
      # Load CSV file
      data = pd.read_csv(filePath)

    if data.isna().any().any() == False:
        print("✅ لايوجد أية بيانات مفقودة")
    else:
        completeness = completeness - ((data.isnull().sum().sum() / data.count().sum()) * 100)
    print("درجة اكتمال البيانات: " + str(completeness))
    return completeness
    print("---------------------------------")

def availability():
  print("---------------------------------")
  print("_____مقياس توفر البيانات وقابليتها للقراءة_____")
  # Availabilty score variable
  availabilityScore = 0
  numberOfReadableFormats= 0
  # Asks the user to enter the number of dataset file formats exist
  numberOfFormats = int(input("الرجاء إدخال عدد تنسيقات الملف الموجودة: "))
  # For loop that is repeated many times as the number of formats exist
  for i in range(numberOfFormats):
        # Asks the user to enter the file path
        dataset_path = input(f"🗎 أدخل مسار التنسيق {i+1} للملف هنا: ")
        try:
            # Open the file exist in the path entered by the user
            with open(dataset_path, 'r') as file:
                # Check whther the file is readable
                file.read()
                # Then increase the availabilty score
                availabilityScore += (1/numberOfFormats)*100
                numberOfReadableFormats +=1
        # If the file didn't open, nothing will happen
        except:
          pass
  # If the number of dataset available formats is equal 2
  if numberOfFormats==2:
      availabilityScore = availabilityScore*(2/3)
      print("❌ لم يتم تجاوز الحد الأدنى (3) من عدد التنسيقات ")
      # If the number of dataset available formats is equal 1
  elif numberOfFormats==1:
      availabilityScore = availabilityScore*(1/3)
      print(" ❌ لم يتم تجاوز الحد الأدنى (3) من عدد التنسيقات ")
  else:
      print("✅ تم تجاوز الحد الأدنى (3) من عدد التنسيقات ")
  # The score will be printed
  availabilityScore= round(availabilityScore, 2)
  print(f"يوجد ({numberOfReadableFormats}) نوع قابل للقراءة من أصل ({numberOfFormats}) تنسيقات ")
  print(f"درجة توفر البيانات وقابليتها للقراءة: %{availabilityScore}")
  return availabilityScore
  print("---------------------------------")

import datetime
def findability() :
    print("---------------------------------")
    print("_____مقياس سهولة العثور على البيانات_____")
    st=''
    findabilityScore=10
    #take the title and check if it does exist if it wasn't subtract by 2
    title = input("أدخل عنوان محموعة البيانات: ")
    if title=='':
            st+='لا يوجد عنوان لمجموعة البيانات'
            findabilityScore-=2
    #take the publish date and check if it was fully entered if it wasn't subtract by 2
    print(" أدخل تاريخ جمع البيانات ميلاديًا")
    year = input("السنة: ")
    month = input("الشهر: ")
    day = input("اليوم: ")
    if year=='' or month=='' or day=='':
        st+='لم يتم ادخال تاريخ جمع البيانات لمجموعة البيانات'
        findabilityScore-=2
    elif (not year=='') and (not month=='') and (not day==''):
        pubDate = datetime.date(int(year),int(month),int(day))
    #take the tags and check if their number are more than 3 if it wasn't subtract by 2 for each missing tag
    tags = []
    count = 0
    tagnum = int(input("أدخل عدد الأوسمة: "))
    if tagnum==2:
        st+='لا يوجد سوى وسمين لمجموعة البيانات'
        findabilityScore-=2
    elif tagnum==1:
        st+='لا يوجد سوى وسم واحد لمجموعة البيانات'
        findabilityScore-=4
    elif tagnum==0:
        st+='لا يوجد وسوم لمجموعة البيانات'
        findabilityScore-=6

    while count<tagnum:
        tag = input("أدخل الأوسمة مبتدءً بـ # قبل ذكر الوسم: ")
        if ('#' in tag):
            count+=1
            tags.append(tag)
        else:
            print("⚠️ لقد نسيت وضع # قبل الوسم")


    print(st)
    print(f" درجة سهولة العثور على البيانات: %{findabilityScore*10}")
    return findabilityScore*10
    print("---------------------------------")


import pandas as pd
def consistency():
    print("---------------------------------")
    print("_____مقياس اتباع النمط (الاتساق)")
    # Asks the user to enter the path of the file
    filePath = input("🗎 الرجاء إدخال مسار الملف: ")
    if filePath.endswith('.xlsx'):
      # Load XLSX file
      dataFrame = pd.read_excel(filePath)
    elif filePath.endswith('.xls'):
      # Load XLSX file
      dataFrame = pd.read_excel(filePath)
    elif filePath.endswith('.csv'):
      # Load CSV file
      dataFrame = pd.read_csv(filePath)
    # Store the number of columns
    numberOfColumns= len(dataFrame.columns)
    inconsistentColumnsCount = 0
    inconsistentColumnsArray = []
    consistentColumnsCount = 0
    consistentColumnsArray = []
    # Check every column data type
    print(f"نوع البيانات لكل عامود: ")
    for column in dataFrame.columns:
        # Print the data type of the column
        print(f"{column} data type: {dataFrame[column].dtype}")

        # Check if there is more than one data type in the same column
        if len(set(dataFrame[column].apply(type))) > 1:
            inconsistentColumnsCount += 1
            inconsistentColumnsArray.append(column)
        else:
          consistentColumnsCount += 1
          consistentColumnsArray.append(column)
        # Compute the consistency score
        consistencyScore= 1- (inconsistentColumnsCount/numberOfColumns)
    if  inconsistentColumnsCount>0:
      print(f"❌الأعمدة التي لا تتبع النمط: {inconsistentColumnsArray}")
    if  consistentColumnsCount>0:
      print(f"✅الأعمدة التي تتبع النمط: {consistentColumnsArray}")
    print(f"درجة اتباع النمط: %{consistencyScore*100}")
    print("---------------------------------")
    return consistencyScore*100
def overallQuality(comp, ava, find, cons):
  quality= ((0.5*comp + 0.4*ava + 0.3*find + 0.4*cons)/160)*100
  print("% درجة الجودة والنضج الكلية: ", quality,)
  print("---------------------------------")

print("______________________________________ ODAF tool ___________________________________________")
print("-----------------| (قطاع الزراعة والأسماك)أداة تعمل على قياس جودة ونضج البيانات المفتوحة |---------------------")
print("                                                                    :القائمة ")
print("1- اكتمال البيانات")
print("2- توفر البيانات وقابليتها للقراءة ")
print("3- سهولة العثور على البيانات ")
print("4- اتباع النمط (الاتساق) ")
print("5- الجودة الكلية ")
print("6- إيقاف تشغيل الأداة ")

print("___________________________________ ")

while True:
  choice = int(input("رقم المعيار المراد قياسه: "))
  if choice ==1:
    completeness()
  elif choice ==2:
    availability()
  elif choice ==3:
    findability()
  elif choice ==4:
    consistency()
  elif choice ==5:
    comp= completeness()
    ava= availability()
    find= findability()
    cons= consistency()
    overallQuality(comp, ava, find, cons)
  elif choice ==6:
    break
  else:
    print("⚠️ الرجاء إدخال رقم صحيح")
```
